## Scatter & Bubble Charts: Gapminder

> **Push to:** `week04/lecture04_exercise.ipynb`

**Rules:**
1. Colour used **sparingly** — one categorical variable, no rainbow
2. If showing all continents, either use accessible palette OR grey all + highlight one
3. `size_max` set when using bubble size
4. Log scale for GDP per capita
5. Insight title

---


```python
import pandas as pd
import plotly.express as px


# Dataset: Gapminder — GDP, Life Expectancy, Population by Country
# Source: Gapminder Foundation (gapminder.org)

df = px.data.gapminder()
print(f"Loaded: {len(df)} rows")
print(df.head())
```

```python
# explore

print(df.info())
print("Years:", sorted(df['year'].unique()))
print("Continents:", df['continent'].unique())
print(df.describe().round(1))
```

## Task 1 — Scatter: life expectancy change over time

**What to build:** A scatter showing **GDP per capita vs life expectancy** for **two years** (2000 and 2007) to show how both moved — use **colour for year** (just 2 colours), **one continent only**.

Choose any continent except Africa (that was the example). Highlight the change direction.

> 💡 Filter: `df.loc[df['continent'] == 'YOUR_CHOICE']` then filter years


```python
df_europe = df.loc[
    (df['continent'] == 'Europe') & (df['year'].isin([1952, 2007]))
].copy()
df_europe['year'] = df_europe['year'].astype(str)

fig1 = px.scatter(
    df_europe,
    x='gdpPercap',
    y='lifeExp',
    color='year',
    color_discrete_map={'1952': '#B25734', '2007': '#038888'},
    hover_name='country',
    log_x=True,
    labels={
        'gdpPercap': 'GDP per Capita (log scale, USD)',
        'lifeExp': 'Life Expectancy (years)',
        'year': 'Year'
    },
    title='European countries increased in welath and life expectancy between 1952 and 2007'
)

fig1.update_traces(marker=dict(size=9, opacity=0.85, line=dict(width=0.5, color='white')))
fig1.update_layout(
    plot_bgcolor='white',
    paper_bgcolor='white',
    font_family='Arial',
    title_font_size=15,
    legend_title_text='Year',
    xaxis=dict(showgrid=True, gridcolor='#eeeeee', zeroline=False),
    yaxis=dict(showgrid=True, gridcolor='#eeeeee', zeroline=False),
    height=500,
)

fig1.show()
```

## Task 2 — Bubble chart: tell a story

**What to build:** A bubble chart (full 2007 dataset, all countries) where:
- x = GDP per capita (log scale)
- y = life expectancy
- size = population
- colour = ONE continent highlighted (your choice), all others grey
- At least one annotation explaining the highlighted group's story

> This is the grey-and-highlight technique applied to a bubble chart.


```python
df_2007 = df.loc[df['year'] == 2007].copy()

HIGHLIGHT = 'Europe'
df_2007['color_group'] = df_2007['continent'].apply(
    lambda c: c if c == HIGHLIGHT else 'Other'
)

df_2007['pop_norm'] = df_2007['pop'] / df_2007['pop'].max() * 100

df_2007 = df_2007.sort_values('color_group', ascending=False)

fig2 = px.scatter(
    df_2007,
    x='gdpPercap',
    y='lifeExp',
    size='pop_norm',
    color='color_group',
    color_discrete_map={HIGHLIGHT: '#038888', 'Other': '#cccccc'},
    size_max=60,
    hover_name='country',
    hover_data={'pop': True, 'pop_norm': False, 'gdpPercap': True, 'lifeExp': True},
    log_x=True,
    labels={
        'gdpPercap': 'GDP per Capita (log scale, USD)',
        'lifeExp': 'Life Expectancy (years)',
        'pop': 'Population',
        'color_group': 'Region'
    },
    title='Europe in 2007 is leading in wealth and life expectancy'
)

fig2.update_traces(
    selector=dict(name='Other'),
    marker=dict(opacity=0.35, line=dict(width=0, color='#cccccc'))
)
fig2.update_traces(
    selector=dict(name=HIGHLIGHT),
    marker=dict(opacity=0.75, line=dict(width=0.5, color='white'))
)

fig2.add_annotation(
    x=4.7, y=80.1,
    text="Norway",
    showarrow=True,
    arrowhead=3,
    ax=80, ay=-40,
    font=dict(size=12, color='#038888'),
    bgcolor='white',
    bordercolor='#038888',
    borderwidth=1,
    borderpad=4,
)

fig2.update_layout(
    plot_bgcolor='white',
    paper_bgcolor='white',
    font_family='Arial',
    title_font_size=15,
    legend_title_text='',
    xaxis=dict(showgrid=True, gridcolor='#eeeeee', zeroline=False),
    yaxis=dict(showgrid=True, gridcolor='#eeeeee', zeroline=False),
    height=550,
)

fig2.show()
```
